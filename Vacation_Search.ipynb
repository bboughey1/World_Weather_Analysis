```python
import pandas as pd
import requests
import gmaps

from config import g_key
```

```python
#import city data
city_data_df = pd.read_csv("data/WeatherPy_Database.csv")
city_data_df
```

```python
city_data_df.dtypes
```

```python
# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# Enter max Temperature criteria
min_temp = float(input("What is the minimum temperature you would like for vacation?"))
max_temp = float(input("What is the maximum temperature you would like for vacation?"))

# Ask the customer about rain
rain_amount = input("Do you want it to be raining? (yes/no)")

# Ask about snow
snow_amount = input("Do you want it to be snowing? (yes/no)")
```

```python
# Set up the filters to provide answers to the input questions from above

if rain_amount == "no" and snow_amount == "no":
    filtered_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp)& 
                                          (city_data_df["Max Temp"] >= min_temp)& 
                                          (city_data_df["Rain (inches)"] == 0) & 
                                            (city_data_df["Snow (inches)"] == 0)]
elif rain_amount == "no" and snow_amount == "yes":
    filtered_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp)&
                                          (city_data_df["Max Temp"] >= min_temp)&
                                          (city_data_df["Rain (inches)"] == 0) &
                                           (city_data_df["Snow (inches)"] > 0)]
elif rain_amount == "yes" and snow_amount == "no":
    filtered_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp)&
                                          (city_data_df["Max Temp"] >= min_temp)&
                                          (city_data_df["Rain (inches)"] > 0.0) &
                                           (city_data_df["Snow (inches)"] == 0)]
else: 
    filtered_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp)&
                                          (city_data_df["Max Temp"] >= min_temp)&
                                          (city_data_df["Rain (inches)"] > 0.0) &
                                           (city_data_df["Snow (inches)"] > 0)]
```

```python
# Clean the data set
filtered_cities_df = filtered_cities_df.dropna()

filtered_cities_df
```

```python
filtered_cities_df.count()
```

```python
# Begin setting up hotel finder
hotel_df = filtered_cities_df[["City", "Country", "Max Temp", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head()
```

```python
# Set search parameters for hotels near cities with desired weather
params ={
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

# Iterate through
for index, row in hotel_df.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    
    params["location"] = f"{lat},{lng}"
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    hotels = requests.get(base_url, params=params).json()
    
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except IndexError:
        print("Hotel Not Found..skipping..")
```

```python
hotel_df.head()
```

```python
output_data_file ="./data/WeatherPy_Vacation.csv"
hotel_df.to_csv(output_data_file, index_label="City_ID")
```

```python
vacation_df = pd.read_csv("./data/WeatherPy_Vacation.csv")
vacation_df.head()
```

```python
# Add hotel marks to the map
info_box_template = """
<dl>
<dt>Hotel Name</dt>
<dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Hotel Name} at {Max Temp}</dd>
"""

hotel_info = [info_box_template.format(**row) for index, row in vacation_df.iterrows()]
locations = vacation_df[["Lat", "Lng"]]
```

```python
# Determine contents to be shown on map, display the map
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig = gmaps.figure()
fig.add_layer(marker_layer)

fig
```
